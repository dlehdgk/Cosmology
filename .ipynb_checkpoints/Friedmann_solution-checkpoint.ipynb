Cosmology PHYS97015 Imperial College London PS1 Q5 solutions to the Friedmann equations for positive curvature.

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
#params
a0 = 1e20
Om0 = 1.02
H0 = 3.241e-18*0.7
```

```python
#functions
def a(theta):
    return a0*Om0*(1-np.cos(theta))/(2*(Om0-1))
def t(theta):
    return Om0*(theta-np.sin(theta))/(2*H0*(Om0-1)**1.5)
```

```python
theta = np.linspace(0,2*np.pi)
scale = a(theta)
time = t(theta)
```

```python
plt.plot(theta,scale,label= 'a($\\theta$)')
plt.grid()
plt.xlabel('$\\theta$')
plt.ylabel('a')
plt.show()
```

```python
plt.plot(theta,time,label='t($\\theta$)')
plt.grid()
plt.xlabel('$\\theta$')
plt.ylabel('t')
plt.show()
```

```python
plt.plot(time,scale)
plt.grid()
plt.xlabel('t')
plt.ylabel('a')
plt.show()
```
