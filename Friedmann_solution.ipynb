```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
#params
a0 = 1e20
Om0 = 1.02
H0 = 3.241e-18*0.7
```

```python
#functions
def a(theta, curvature):
    if curvature == "Open":
        return -a0*Om0*(1-np.cosh(theta))/(2*(Om0-1))
    elif curvature == "Closed":
        return a0*Om0*(1-np.cos(theta))/(2*(Om0-1))
    
def t(theta, curvature):
    if curvature == "Open":
        return -Om0*(theta-np.sinh(theta))/(2*H0*(Om0-1)**1.5)
    elif curvature == "Closed":
        return Om0*(theta-np.sin(theta))/(2*H0*(Om0-1)**1.5)
```

```python
theta = np.linspace(0,2*np.pi)
sf = np.array([a(theta,"Open"),a(theta,"Closed")])
time = np.array([t(theta,"Open"),t(theta,"Closed")])
```

```python
#scale factor to parameter
fig,ax = plt.subplots(2,1,sharex=True)
for i in range(2):
    ax[i].plot(theta,sf[i,:])
    ax[i].grid()
    ax[i].set_ylabel('a($\\theta$)')
ax[1].set_xlabel('$\\theta$')
ax[0].legend(["Open"])
ax[1].legend(["Closed"])
fig.show()
```

```python
#time to parameter
fig,ax = plt.subplots(2,1,sharex=True)
for i in range(2):
    ax[i].plot(theta,time[i,:])
    ax[i].grid()
    ax[i].set_ylabel('t($\\theta$)')
ax[1].set_xlabel('$\\theta$')
ax[0].legend(["Open"])
ax[1].legend(["Closed"])
fig.show()
```

```python
#scale factor to time
fig,ax = plt.subplots(2,1)
for i in range(2):
    ax[i].plot(time[i,:],sf[i,:])
    ax[i].grid()
    ax[i].set_ylabel('a')
    ax[i].set_ylabel('t')

ax[0].set_title("Open")
ax[1].set_title("Closed")
fig.tight_layout()
fig.show()
```
